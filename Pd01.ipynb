```python
# Importando as libs
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Criando o DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Visualizando as 5 primeiras linhas 
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipos de dados 
df.dtypes
```

```python
# Qual é a Receita Total?
df["Valor Venda"].sum()
```

```python
# Qual é o custo Total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) 
```

```python
df.head()
```

```python
round(df["Custo"].sum(), 2)
```

```python
# Criando coluna com lucro total 
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
# Total do Lucro

round(df["Lucro"].sum() , 2)
```

```python
# Criando uma coluna com total de dias para enviar o produto

df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
# Para saber a média do tempo de envio de cada Marca antes precisamos converter a coluna Tempo_envio para tipo númerico

# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
# Conferindo o tipo
df["Tempo_envio"].dtypes
```

```python
# Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

## Missing Values

```python
# Verificando se temos dados faltantes 
df.isnull().sum()
```

## Descobrindo o Lucro por Ano e Marca


```python
# Agrupando por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Formatando os valores

pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto")
```

```python
# Lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano");
plt.xlabel("Ano");
plt.ylabel("Receita");
```

```python
# Analisando número do lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```
